### 1. How many distinct (different) actors' last names are there?

```python
data = pd.read_sql_query("""SELECT count(DISTINCT(last_name)) AS 'Last names' 
FROM sakila.actor""", engine)
data.head()
```

### 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
data = pd.read_sql_query("""SELECT *, SUBSTR(rental_date, 1,4) AS 'Year', SUBSTR(rental_date, 6,2) AS 'Month', SUBSTR(rental_date, 9, 2) AS 'Weekday' 
FROM sakila.rental""", engine)
display(data.head(10))
display(data.tail(10))
```

### 3. Get all films with ARMAGEDDON in the title.

NOTE: MySLQ only one percentage symbol % for function LIKE; in Python %%  

```python
data = pd.read_sql_query("""SELECT title AS 'Title' 
FROM sakila.film 
WHERE title LIKE '%%ARMAGEDDON%%'""", engine)
data.head()
```

### 4. Get 10 the longest films.

Note: the output is different depending on whether you show 10 first records with SQL or with Python.

```python
# Option 1: Limiting amount of rows in SQL
data= pd.read_sql_query("""SELECT title, length 
FROM sakila.film 
ORDER BY length DESC
LIMIT 10""", engine)
data
```

```python
# Option 2: Limiting amount of rows in Python
data = pd.read_sql_query("""SELECT title, length 
FROM sakila.film 
ORDER BY length DESC""", engine)
data.head(10)
```

### 5. How many films include Behind the Scenes content?

```python
data = pd.read_sql_query("""SELECT count(*) AS 'Behind the scenes' FROM sakila.film
WHERE special_features LIKE '%%Behind the scenes%%'""", engine)
data
```

### 6. Which kind of movies (rating) have a mean duration of more than two hours?

```python
data = pd.read_sql_query("""SELECT round(avg(length), 2) AS 'Mean length', rating AS 'Rating' FROM sakila.film
GROUP BY rating
HAVING AVG(length) > 120""", engine)
data
```

### 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank. 

```python
data = pd.read_sql_query("""SELECT RANK() OVER(ORDER BY length DESC) AS 'Rank', title AS 'Title', length AS 'Length' 
FROM sakila.film
WHERE length > 0 AND length IS NOT NULL
ORDER BY length DESC""", engine)
data
```
